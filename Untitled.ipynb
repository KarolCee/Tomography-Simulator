```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.color import rgb2gray
from skimage import io
from bresenham import bresenham
import math
import ipywidgets
from scipy.fftpack import fft, ifft, fftfreq
```

# WGRANIE ZDJĘCIA WEJŚCIOWEGO (uzywanego jako zmienna globalna)

```python
image = io.imread('logan_small.jpg')
image = rgb2gray(image)

#wyznaczenie centralnego punktu zdjecia - osi obrotu skanowania, wyznaczenie promienia okr. wpisanego
x_c = image.shape[0]//2
y_c = image.shape[1]//2
r = max(image.shape)//2
```

# FUNKCJA TWORZĄCA SINOGRAM (TRANSFORMATA RADONA)

```python
def radon(n=1,rozpietosc=180,alpha=1,kat_stopu=180,iteracyjnie=False,filtruj=False):
    ims = []
    
    alfas = np.arange(0,kat_stopu,alpha)
    theta = rozpietosc / (n - 1)  # kat przypadajacy jednemu emiterowi (stopnie), odchylka od alfy dla glownej lini rowna 0
    R = np.zeros((n,len(alfas))) #radon image
    
    #step = image.shape[0]/n

    
    linie_globalnie = []
    sumy_globalnie = []
    for iter,alfa in enumerate(alfas):

        emitery = []
        detektory = []

        
        #KĄTY ODCHYŁEK LINII EMITER-DETEKTOR
        start_val = math.floor(n/2) 
        desc = True
        for i in range(n): #dla kazdej iteracji
            if desc: #pierwsza polowa emiterow nad emiterem srodkowym
                angle = (start_val*theta+alfa) #emiter przesuniety o krotnosc theta
                start_val = start_val - 1
                if start_val == 0:
                    desc = False
            else: #druga polowa emiterow pod emiterem srodkowym
                angle = (-start_val*theta+alfa) #emiter przesuniety o krotnosc theta
                start_val = start_val + 1

        #obliczenie wspolrzednych emiterow i detektorow na okregu
            if angle == 270.0:
                x = x_c
                y = y_c - r
            elif angle > 180:
                angle = angle % 180
                angle = np.deg2rad(angle)
                y = y_c - r * math.sin(angle)
                x = x_c - r * math.cos(angle)
            elif angle == 180:
                x = x_c - r
                y = y_c
            elif angle > 90:
                angle = angle % 90
                angle = 90 - angle
                angle = np.deg2rad(angle)
                y = y_c + r * math.sin(angle)
                x = x_c - r * math.cos(angle)
            elif angle == 90:
                y = y_c + r
                x = x_c
            elif angle > 0:
                angle = np.deg2rad(angle)
                y = y_c + r * math.sin(angle)
                x = x_c + r * math.cos(angle)
            elif angle == 0:
                x = x_c + r
                y = y_c
            elif angle == -90:
                x = x_c
                y = y_c - r
            elif angle < 0:
                angle = abs(angle)
                angle = np.deg2rad(angle)
                y = y_c - r * math.sin(angle)
                x = x_c + r * math.cos(angle)
            x_e = round(x)
            y_e = round(y)
            x = x_c - x_e
            y = y_c - y_e
            x_d = x_c + x
            y_d = y_c + y
            x_d = round(x_d)
            y_d = round(y_d)
            detektory.append([x_d,y_d])
            emitery.append([x_e,y_e])

        detektory.reverse()
        linie = []
        sumy = []
        for nr_linii,id in enumerate(zip(emitery,detektory)):
            i = id[0]
            d = id[1]
            x_e = i[0]
            y_e = i[1]
            x_d = d[0]
            y_d = d[1]
            linia = list(bresenham(x_e,y_e,x_d,y_d))
            linie.append(linia)
            dosumy = []
            for piksel in linia:
                x = piksel[0]
                y = abs(len(image) - piksel[1])
                if x>=image.shape[1]:
                    x = image.shape[1]-1
                if y>=image.shape[0]:
                    y=image.shape[0]-1
                kolor = image[y][x]
                dosumy.append(kolor)
            sumka = sum(dosumy)/len(linia)
            R[round(nr_linii),iter] = sumka
            sumy.append(sumka)
        if iteracyjnie:
            ims.append(R.copy())
        sumy_globalnie.append(sumy)
        
        linie_globalnie.append(linie)
        
        
       
    if filtruj:
        filtered_R = R.copy() #kopiowanie sinogramu
        f = fftfreq(filtered_R.shape[0]).reshape(-1, 1) 
        projection = fft(filtered_R,axis=0) * np.abs(f)
        filtered_R = np.real(ifft(projection, axis=0))
        return filtered_R,R,sumy_globalnie,linie_globalnie,ims
    return [],R,sumy_globalnie,linie_globalnie,ims
```

# FUNKCJA ODWTARZAJĄCA OBRAZ Z SINOGRAMU (ODWROTNA TRANSFORMATA RADONA)

```python
def inverse_radon(sinogram,sumy_globalnie,linie_globalnie,iteracyjnie=False,filtruj=False):
    ims = []
    IR = np.zeros(image.shape)
    
    
    if filtruj:
        sumy_globalnie = []
        for row in sinogram.T:
            sumki = []
            for piksel in row:
                sumki.append(piksel)
            sumy_globalnie.append(sumki)

    for lines_of_angle,sums_of_angle in zip(linie_globalnie,sumy_globalnie):
        for line,suma in zip(lines_of_angle,sums_of_angle):
            for point in line:
                x = point[0]
                y = abs(len(image) - point[1])
                if x>=image.shape[1]:
                    x = image.shape[1]-1
                if y>=image.shape[0]:
                    y=image.shape[0]-1
                IR[y][x] = IR[y][x]+suma 
        if iteracyjnie:
            ims.append(IR.copy())
    return IR/IR.max(),ims #dzielimy przez Max w celu normalizacji 
```

# FUNKCJA DO ZAPISU OBRAZU W FORMACIE DICOM

```python
import pydicom
from pydicom.data import get_testdata_files

def write_dicom(image,nazwa_pacjenta,data,komentarz,nazwa_pliku):
    filename = get_testdata_files('CT_small.dcm')[0]
    dataset = pydicom.dcmread(filename)
    
    dataset.PatientName = nazwa_pacjenta
    dataset.StudyDate = data
    dataset.StudyDescription = komentarz
    dataset.Rows = image.shape[0]
    dataset.Columns = image.shape[1]
    image = np.sign(image*65535)*(np.abs(image*65535)) **(1/2.2) #**(1/2.2) -> korekcja gamma
    np.round(image,decimals=0,out=image)
    image = image.astype(np.uint16)
    

    dataset.PixelData = image.tostring()
    
    dataset.save_as(nazwa_pliku)
```

# FUNKCJA DO ODCZYTU FORMATU DICOM

```python
def read_dicom(filename):
    dataset = pydicom.dcmread(filename)
    print("Patient's name...:", dataset.PatientName)
    print("Comments.........:", dataset.StudyDescription)
    print("Study Date.......:", dataset.StudyDate)
    plt.imshow(dataset.pixel_array,cmap=plt.cm.bone)
    plt.show()
```

# FUNKCJA TWORZĄCA SINOGRAM ORAZ OBRAZ WYNIKOWY (Z EWENTUALNYM FILTREM)

```python
import imageio
def animation(ims, filename):
    scale = ims[-1].max()
    gif = list(map(lambda x: (x/scale*255).astype(np.uint8), ims))
    imageio.mimsave(filename+".gif", gif)

from IPython.display import Image
def radon_iradon(maxKatUkladu=180,rozpietosc=180,alpha=3,n=image.shape[0]//4,iteracyjnie=False,filtruj=False):
    sinogram,radon_image,sumy_globalnie,linie_globalnie,radon_gifs = radon(n=n,kat_stopu=maxKatUkladu,rozpietosc=rozpietosc,alpha=alpha,iteracyjnie=iteracyjnie,filtruj=filtruj)
    global inverse_radon_image
    inverse_radon_image,iradon_gifs = inverse_radon(sinogram,sumy_globalnie,linie_globalnie,iteracyjnie=iteracyjnie,filtruj=filtruj)
    
    
    if iteracyjnie:
        plt.imshow(image,cmap=plt.cm.gray)
        plt.show()
        animation(radon_gifs,'radonGI')
        animation(iradon_gifs,'iradonGI')
        display(Image("radonGI.gif"))
        display(Image("iradonGI.gif"))
    else:
        plt.title("obraz wejściowy")
        plt.ylabel("y")
        plt.xlabel("x")
        plt.imshow(image,cmap=plt.cm.gray)
        plt.show()
        plt.title("SINOGRAM (TRANSFORMATA RADONA)")
        plt.ylabel("pary emiter-detektor")
        plt.xlabel("kolejne kąty")
        plt.imshow(radon_image,cmap=plt.cm.gray)
        plt.show()
        plt.title("ODWROTNA TRANSFORMATA RADONA (OBRAZ WYJŚCIOWY)")
        plt.ylabel("y")
        plt.xlabel("x")
        plt.imshow(inverse_radon_image,cmap=plt.cm.gray)
        plt.show()
```
